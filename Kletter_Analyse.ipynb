```python
import os
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
data = []
for file in os.listdir():
  if "Kletter-Tagebuch" in file:
    print(file)
    data.append(pd.read_csv(file, decimal=","))
```

```python
frame = pd.concat(data, axis=1, ignore_index=True)
frame.drop([13,24], axis="rows", inplace=True)
frame.drop([32, 64, 96, 128, 160, 192, 224], axis="columns", inplace=True)
```

```python
df = frame.T.set_axis(frame[0], axis="columns")
df.drop(0, axis="rows", inplace=True)
```
